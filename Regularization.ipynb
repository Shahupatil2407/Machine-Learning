```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Advertising.csv")
```

```python
df
```

```python
X = df.drop('sales', axis=1)
y = df['sales']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_converter = PolynomialFeatures(degree=3, include_bias=False)
```

```python
poly_features = poly_converter.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
X_train[0]
```

```python
X_test[0]
```

***Ridge Regression ***

```python
from sklearn.linear_model import Ridge
```

```python
model = Ridge(alpha=10)
```

```python
model.fit(X_train, y_train)
```

```python
predictedions = model.predict(X_test)
```

```python
predictedions
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE = mean_absolute_error(y_test, predictedions)
MAE
```

```python
MSE = mean_squared_error(y_test, predictedions)
MSE
```

```python
RMSE = np.sqrt(MSE)
RMSE
```

```python
from sklearn.linear_model import RidgeCV
```

```python
ridge_cv_model = RidgeCV(alphas=(0.1, 1.0, 10.0), scoring='neg_mean_absolute_error')
```

```python
ridge_cv_model.fit(X_train, y_train)
```

```python
ridge_cv_model.alpha_
```

```python
test_prediction = ridge_cv_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test, test_prediction)
MAE
```

```python
MSE = mean_squared_error(y_test, test_prediction)
MSE
```

```python
RMSE = np.sqrt(MSE)
RMSE
```

**Lasso Regression** - Least Absolute Shrinkage and Selection Operator

```python
from sklearn.linear_model import LassoCV
```

```python
lasso_cv_model = LassoCV(eps=0.01,n_alphas=100,cv=5, max_iter=1000000)
```

```python
lasso_cv_model.fit(X_train, y_train)
```

```python
lasso_cv_model.alpha_
```

```python
predictions = lasso_cv_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test, predictions)
MAE
```

```python
MSE = mean_squared_error(y_test, predictions)
MSE
```

```python
RMSE = np.sqrt(MSE)
RMSE
```

```python
lasso_cv_model.coef_
```

**Elastic Net**

```python
from sklearn.linear_model import ElasticNetCV
```

```python
elastic_model = ElasticNetCV(l1_ratio=[.1, .5, .7, .9, .95, .99, 1], eps = 0.001, n_alphas=100, max_iter=1000000)
```

```python
elastic_model.fit(X_train, y_train)
```

```python
model_prediction = elastic_model.predict(X_test)
```

```python
model_prediction
```

```python
MAE = mean_absolute_error(y_test, model_prediction)
MAE
```

```python
MSE = mean_squared_error(y_test, model_prediction)
print("Mean Squared Error:", MSE)
```

```python
RMSE = np.sqrt(MSE)
RMSE
```

```python
elastic_model.coef_
```

```python
elastic_model.l1_ratio_
```
